```python
# # Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
# Ans =Seaborn is a popular data visualization library in Python that provides a wide range of plotting functions to create 
# informative and visually appealing visualizations. Here are five common plots that can be created using Seaborn along with their uses:
# Scatter Plot - A scatter plot is used to display the relationship between two variables.
# Seaborn's scatter plot function scatterplot() is used to create scatter plots. Scatter plots are useful for identifying patterns and trends in data, identifying outliers and visualizing the spread of the data.
# Line Plot - A line plot is used to show how a variable changes over time. 
# Seaborn's line plot function lineplot() is used to create line plots. Line plots are useful for visualizing trends,
# identifying patterns, and seeing how a variable changes over time.
# Bar Plot - A bar plot is used to compare categorical variables.
# Seaborn's bar plot function barplot() is used to create bar plots. Bar plots are useful for comparing the frequency,
# mean or median of a categorical variable across different groups.
# Heat Map - A heat map is used to visualize data using color-coded cells.
# Seaborn's heat map function heatmap() is used to create heat maps. Heat maps are useful for identifying patterns
# and trends in large datasets, identifying correlations and dependencies between variables, and visualizing matrices and tables.
# Box Plot - A box plot is used to summarize the distribution of a variable.
# Seaborn's box plot function boxplot() is used to create box plots. Box plots are useful for identifying outliers,
# visualizing the spread and shape of the data, and comparing the distribution of a variable across different groups.
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.
import seaborn as sns 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load fmri dataset
fmri = sns.load_dataset("fmri")

# plot line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

# set axis labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal by Timepoint, Event, and Region")
plt.show()
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
# 'pclass', y = 'age' and y = 'fare'.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load titanic dataset
titanic = sns.load_dataset("titanic")

# plot box plots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
sns.boxplot(x='pclass', y='age', data=titanic, ax=axes[0])
sns.boxplot(x='pclass', y='fare', data=titanic, ax=axes[1])

# set axis labels and title
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Age')
axes[0].set_title('Age by Passenger Class')

axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Fare')
axes[1].set_title('Fare by Passenger Class')

plt.show()
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.
import seaborn as sns
import matplotlib.pyplot as plt

# load diamonds dataset
diamonds = sns.load_dataset('diamonds')
# # plot histogram of diamond carat weights
sns.histplot(data=diamonds, x='carat', bins=30, kde=True)

# set axis labels and title
plt.xlabel('Carat Weight')
plt.ylabel('Frequency')
plt.title('Histogram of Diamond Carat Weights')

plt.show()
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.
# Ans 5
import seaborn as sns

# load iris dataset
iris = sns.load_dataset('iris')

# plot pair plot
sns.pairplot(iris, hue='species')
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
# load flights dataset
flights = sns.load_dataset('flights')

# reshape the data to wide format
flights_wide = flights.pivot(index='month', columns='year', values='passengers')

# plot heatmap
sns.heatmap(flights_wide, cmap='YlGnBu')

# set axis labels and title
sns.set(font_scale=1.2)
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Count by Year and Month')

plt.show()
```
